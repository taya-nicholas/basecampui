```
#| export
def AIUserMessage(message:str, user="User", cls_top="text-xs", cls_mid="text-sm"):
    return Div(
        Div(
            Muted(user, cls=cls_top),
            Muted("•", cls=cls_top),
            Muted("5:24 PM", cls=cls_top, **tooltip_kwargs("Apr 14, 5:24 PM")),
            cls="flex gap-2"
        ),
        Div(
            render_md(message),
            cls=f"flex flex-col gap-4 {cls_mid}"
        ),
        cls="flex flex-col border border-transparent hover:border-border rounded-md pt-7 pb-7 pl-3 pr-6 whitespace-pre-wrap gap-2"
    )
```

```
@rt
def index():
    return Div(
        Div(
            ToggleButton(),
            H4("SAM3: Refining Low-Res Predictions"),
            cls="flex p-3 gap-3 items-center",
        ),
        Div(
            Div(
                AIUserMessage(user_m),
                AIUserMessage(model_m, user="Model"),
                cls="flex flex-col pt-7 max-w-5xl mx-auto pl-4 pr-1 gap-3",
            ),
            cls="overflow-y-auto",
        ),
        Div(
            Div(
                Textarea(placeholder="Start typing a prompt to see what our models can do", type="text", cls="bg-transparent border-none w-full focus:outline-none focus:ring-0 resize-none max-h-32 overflow-y-auto pb-0 min-h-8",
                    oninput="document.getElementById('run-btn').disabled = this.value.trim() === ''",
                ),
                Div(
                    Button("Run Ctrl ↵", id="run-btn", disabled=True, cls=ButtonT.outline + "rounded-xl"),
                    cls="flex justify-end pr-4 pt-0 mt-0",
                ),
                cls="bg-card rounded-lg mb-6 border border-1 pb-4 ml-5 mr-2",
            ),
            cls="w-5xl mx-auto"
        ),
        cls="w-full h-screen flex flex-col"
    )
```
