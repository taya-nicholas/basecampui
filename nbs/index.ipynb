# BasecampUI

BasecampUI is a FastHTML UI library built on top of the excellent Basecoat. It's design goal is to create complex UI components with minimal boilerplate, while still allowing as much customization as possible

## Quickstart
```sh
pip install basecampui
```
```python
from fasthtml.common import *
from basecampui.all import *

app = fastHTML()
rt = app.route

@rt
def index():
    return Div(
        Breadcrumb(['Home', 'Documents', 'Doc 1']),
    )

serve()
```

## Using Icons
BasecampUI uses fastlucide for icons. To make things simple, `basecampui.all` provides a standalone `Icon` function which accepts any lucide icon string. This function creates a new SVG for each icon instance making it easy to quickly test and change for development purposes.

```python
Div(
    Icon("apple")
)
```

If you are happy with your selection of icons, you can include them in your app definition and switch to using the spritesheet singleton. This is more efficient for rendering multiple of the same icon, and allows you to use the rest of the fastlucide features.

```python
app = FastHTML(icons=["apple"])

Div(
    spritesheet("apple")
)
```

Code highlighting is included in `ChatCode` and `CopyableCode`. 
In order to activate, you must return the `CodeHighlightThemeScript()`. You can also pass in custom synatx to this, and include in the footers automatically like so:

```python
ftrs = CodeHighlightThemeScript("FastHTML my_custom_func CustomClass2")
app = FastHTML(ftrs=ftrs)
```
