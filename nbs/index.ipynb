## Developer Guide

If you are new to using `nbdev` here are some useful pointers to get you started.

### Install fhbasecoat in Development mode

```sh
# make sure fhbasecoat package is installed in development mode
$ pip install -e .

# make changes under nbs/ directory
# ...

# compile to have changes apply to fhbasecoat
$ nbdev_prepare
```

## Usage

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/taya-nicholas/fhbasecoat.git
```

or from [conda][conda]

```sh
$ conda install -c taya-nicholas fhbasecoat
```

or from [pypi][pypi]


```sh
$ pip install fhbasecoat
```
